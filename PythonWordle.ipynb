## Function to check if the Word is Valid

```python
def is_word(word):
    try:
        output = dictionary.meaning(word, disable_errors=True)
    except:
        return False
    else:
        return bool(output)
```

## Initializing Variables and Lists

```python
word = data.sample().iloc[0]["Word"][1:]
keyboard = {"q":"","w":"","e":"","r":"","t":"","y":"","u":"","i":"","o":"","p":"","a":"","s":"","d":"","f":"","g":"","h":"","j":"","k":"","l":"","z":"","x":"","c":"","v":"","b":"","n":"","m":""}
fin = False
grid_memory = []
col={}
user_list = []
```

## Function to Assign Colors to a Single Row of the Grid

```python
def assign_colors_grid(word, user_input, col, keyboard):
    col = {0:"", 1:"", 2:"", 3:"", 4:""}
    for i in range(5):
        if(user_input[i] not in word):
            col[i] = "on_grey"
            keyboard[user_input[i]] = "on_grey"

    for i in range(5):
        if(user_input[i] == word[i]):
            col[i] = "on_green"
            keyboard[user_input[i]] = "on_green"

    for i in range(5):
        if(col[i] == "on_green" or col[i] == "on_grey"):
            continue
        count = 0
        for j in range(5):
            if(user_input[j] == user_input[i] and col[j] != "on_green"):
                count += 1
        let_count = word.count(user_input[i])
        j = 0 
        while(count > 0 and let_count > 0):
            if(user_input[j] == user_input[i] and col[j] == "" ):
                col[j] = "on_yellow"
                let_count -= 1
                count -= 1
            j += 1
            if(j == 5):
                break
        for j in range(5):
            if(user_input[j] == user_input[i] and col[j] == ""):
                col[j] = "on_grey"
        
        if(user_input[i] != word[i] and user_input[i] in word):
            if(keyboard[user_input[i]] == "on_green"):
                continue
            else:
                keyboard[user_input[i]] = "on_yellow"

        for i in range(5):
            if(col[i] == "on_grey"):
                keyboard[user_input[i]] = "on_grey"

    grid_memory.append(col)
    return col
```

## Function to Print Keyboard

```python
def print_keyboard(keyboard):
    alph_list = list(keyboard)
    print("     Keyboard")
    for i in range(26):
        if(i == 10 or i == 19):
            print("\n")
        if(keyboard[alph_list[i]] != ''):
            print((colored(alph_list[i], "white", keyboard[alph_list[i]])), end=" ")
        else:
            print(alph_list[i], end=" ")
```

## Function to Print the Whole Grid

```python
def print_full_grid(i, grid_memory, user_list):
    for k in range(i+1):
        for m in range(5):
            print(colored(user_list[k][m], "white", grid_memory[k][m]), end=" ")
        print("\n")
    for k in range(5-i):
        print("_ _ _ _ _")
```

## Function to Print the Colored Squares Grid

```python
def print_squares(x, grid_memory):
    for i in range(x):
        #print("\n")
        for j in range(5):
            if(grid_memory[i][j] == "on_green"):
                print("ðŸŸ©", end="")
            elif(grid_memory[i][j] == "on_yellow"):
                print("ðŸŸ¨", end="")
            elif(grid_memory[i][j] == "on_grey"):
                print("â¬›", end="")
            if(j == 4):
                print("\n")
```
